```python
import matplotlib.pyplot as plt

# Plot instellingen
plt.figure(figsize=(10, 6))

# Lijnen tekenen voor elk model
plt.plot(model_data1['Aantal gekochte EV'], label='Model 1')
plt.plot(model_data2['Aantal gekochte EV'], label='Model 2')
plt.plot(model_data3['Aantal gekochte EV'], label='Model 3')

# Grafiek aankleden
plt.title('Verkoop van EV over tijd')
plt.xlabel('Tijd (stappen)')
plt.ylabel('Aantal gekochte EVs')
plt.legend()  # Voegt de legenda toe
plt.grid(True)

# Toon de grafiek
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2)

x = range(len(model_data1))

# Eerste subplot 
axs[0, 0].plot(x, model_data1['Gemiddelde belangstelling'], label='Model 1')
axs[0, 0].plot(x, model_data2['Gemiddelde belangstelling'], label='Model 2')
axs[0, 0].plot(x, model_data3['Gemiddelde belangstelling'], label='Model 3')

axs[0, 0].set_title("Gemiddelde belangstelling")
axs[0, 0].legend()

# Tweede subplot
axs[0, 1].plot(x, model_data1["Percerntage EV's van Auto's"], label='Model 1')
axs[0, 1].plot(x, model_data2["Percerntage EV's van Auto's"], label='Model 2')
axs[0, 1].plot(x, model_data3["Percerntage EV's van Auto's"], label='Model 3')

axs[0, 1].set_title("Percentage EV van alle auto's")
axs[0, 1].legend()

# Derde subplot
axs[1, 0].plot(x, model_data1["Percentage late majority met EV"], label='Model 1')
axs[1, 0].plot(x, model_data2["Percentage late majority met EV"], label='Model 2')
axs[1, 0].plot(x, model_data3["Percentage late majority met EV"], label='Model 3')

axs[1, 0].set_title('Percentage late majority met EV')
axs[1, 0].legend()

# Vierde subplot 
axs[1, 1].plot(x, model_data1["Hoeveelheid Subsidie"], label='Model 1')
axs[1, 1].plot(x, model_data2["Hoeveelheid Subsidie"], label='Model 2')
axs[1, 1].plot(x, model_data3["Hoeveelheid Subsidie"], label='Model 3')

axs[1, 1].set_title('Totale hoeveelheid subsidie uitgegeven')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
