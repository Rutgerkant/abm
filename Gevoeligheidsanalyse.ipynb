```python
from BaseModel import BaseModelSub
from SubsidieModel import SubsidieModel
from SubsidieModelLinear import SubsidieModel2
from SubsidieModel3 import SubsidieModel3

import random
import numpy as np
import matplotlib.pyplot as plt

initial_waardes_prijs_ev = [32800, 41000, 49200]
initial_waardes_prijs_fba = [26400, 33000, 39600]
initial_waardes_drempelleeftijd = [38, 48, 58]

initial_waardes_gevoeligheidsanalyse = [32800, 41000, 49200]

avg_percentage_ev = np.zeros((3,20))

for j, i in enumerate(initial_waardes_gevoeligheidsanalyse):
    for iteratie in range(20):
        model1 = SubsidieModel(i)
        for k in range(192):
            model1.step()
        
        data1 = model1.datacollector.get_model_vars_dataframe()
        avg_percentage_ev[j, iteratie]= data1["Percerntage EV's van Auto's"].iloc[-1]
    

x = initial_waardes_gevoeligheidsanalyse

fig, ax = plt.subplots()
for i in range(20):
    ax.scatter(x, avg_percentage_ev[:, i], s=50, label=f'Iteration {i+1}')


ax.set_ylim(0.70, ax.get_ylim()[1])  # Set the lower limit of the y-axis to 2300
plt.show()


som_neg = 0
som_pos = 0
som_bas = 0
print("Gemiddelde percentage EV's voor elke iteratie:")
for i in range(20):

    for j in range(3):
        if j == 0:
            som_neg += avg_percentage_ev[j][i]

        elif j == 1:
            som_bas += avg_percentage_ev[j][i]
        elif j == 2:
            som_pos += avg_percentage_ev[j][i]
        
        print(f"Iteratie {i+1}: {avg_percentage_ev[:, i]}")

avg_neg = som_neg/20
avg_pos =som_pos/20
avg_bas = som_bas/20
    
    
print(f"-10 {avg_neg}")
print(f"Basis {avg_bas}")
print(f"+10 {avg_pos}")
```
