```python
from BaseModel import BaseModelSub
from SubsidieModel import SubsidieModel
from SubsidieModelLinear import SubsidieModel2
from SubsidieModel3 import SubsidieModel3

import random
import numpy as np
import matplotlib.pyplot as plt

initial_waardes_gevoeligheidsanalyse = [32800, 41000, 49200]


avg_percentage_ev = np.zeros((3,10))

for j, i in enumerate(initial_waardes_gevoeligheidsanalyse):
    for iteratie in range(10):
        model1 = SubsidieModel(i)
        for k in range(240):
            model1.step()
        
        data1 = model1.datacollector.get_model_vars_dataframe()
        avg_percentage_ev[j, iteratie]= data1["Percerntage EV's van Auto's"].iloc[-1]
    

x = initial_waardes_gevoeligheidsanalyse

fig, ax = plt.subplots()
for i in range(10):
    ax.scatter(x, avg_percentage_ev[:, i], s=50, label=f'Iteration {i+1}')


ax.set_ylim(0.80, ax.get_ylim()[1])  # Set the lower limit of the y-axis to 2300
ax.legend()
plt.show()

print("Gemiddelde percentage EV's voor elke iteratie:")
for i in range(10):
    print(f"Iteratie {i+1}: {avg_percentage_ev[:, i]}")
```
